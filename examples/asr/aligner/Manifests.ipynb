```python
import ast
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# manifests = ['/home/ebakhturina/NeMo/examples/asr/aligner/output/GTC2019/align/GTC2019-keynote-01.wav_manifest.json']\
    # ,
    #          '/home/ebakhturina/NeMo/examples/asr/aligner/output/GTC2019/align/GTC2019-keynote-01.wav_manifest.json',
    #          '/home/ebakhturina/NeMo/examples/asr/aligner/output/GTC2019/align/GTC2019-keynote-01.wav_manifest.json',
    #          '/home/ebakhturina/NeMo/examples/asr/aligner/output/GTC2019/align/GTC2019-keynote-01.wav_manifest.json']

manifests_files = ['/Users/ebakhturina/data/jhh-samples/GTC2019-keynote-01.wav_manifest.json',
                   '/Users/ebakhturina/data/jhh-samples/GTC2019-keynote-02.wav_manifest.json',
                   '/Users/ebakhturina/data/jhh-samples/GTC2019-keynote-03.wav_manifest.json']

audio_files, durations, texts = [], [], []
for manifest in manifests_files:
    with open(manifest, 'r') as f:
        for line in f:
            line = ast.literal_eval(line)
            audio_files.append(line['audio_filepath'])
            durations.append(line['duration'])
            texts.append(line['text'])

bins = [9, 16.7, 25, 40, 70]
bin_names = ['<' + str(bins[0])] + ['[' + str(bins[i]) + '-' + str(bins[i+1]) + ']' for i in range(len(bins)-1)] + ['>' + str(bins[-1])]
bin_assignment = np.digitize(np.array(durations), bins, right=True)
bin_assignment = Counter(np.digitize(np.array(durations), bins))
```

```python
names, values = zip(*sorted(bin_assignment.items(), key = lambda i: i[0]))
names = [bin_names[k] for k in names]
values = np.round(np.array(values) / sum(values) * 100, 2)

plt.ylabel('% to total')
plt.bar(names, values)
```
